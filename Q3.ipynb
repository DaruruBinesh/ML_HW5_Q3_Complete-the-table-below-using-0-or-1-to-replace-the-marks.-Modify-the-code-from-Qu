```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
bt1 = pd.read_csv("Q3_train_data.csv")
bvs1 = pd.read_csv("Q3_dev_data.csv")
bts1 = pd.read_csv("Q3_test_data.csv")
bps1 = np.array(pd.read_csv("Q3_pred_data.csv"))
```

```python
model = tf.keras.Sequential([
	tf.keras.layers.InputLayer(input_shape=(5), name="InputLayer"),
	tf.keras.layers.Dense(5, activation="relu", name="HiddenLayer.1"),
	tf.keras.layers.Dropout(0.1),
	tf.keras.layers.Dense(5, activation="relu", name="HiddenLayer.2"),
  tf.keras.layers.Dropout(0.1),
  tf.keras.layers.Dense(5, activation="relu", name="HiddenLayer.3"),
	tf.keras.layers.Dense(3, activation="sigmoid", name="OutputLayer")
])
model.compile(optimizer="adam", loss="mse")
model.summary()
```

```python
class LossThresholdCallback(tf.keras.callbacks.Callback):
	def __init__(self, threshold):
		super(LossThresholdCallback, self).__init__()
		self.threshold = threshold
	def on_epoch_end(self, epoch, logs=None):
		if logs['val_loss'] <= self.threshold:
			self.model.stop_training=True

print("TRAINING ML PROGRAM")
model.fit(
	x=np.array(bt1.loc[:,"X4":"X0"]),
	y=np.array(bt1.loc[:,"Y2":"Y0"]),
	validation_data=(
		np.array(bvs1.loc[:,"X4":"X0"]),
		np.array(bvs1.loc[:,"Y2":"Y0"])
	),
	batch_size=8,

	epochs=300,
	callbacks=[LossThresholdCallback(0.03)]
)
```

```python
print("TESTING ML PROGRAM")
model.evaluate(
	x=np.array(bts1.loc[:,"X4":"X0"]),
	y=np.array(bts1.loc[:,"Y2":"Y0"])
)
```

```python
print("PREDICTION ML PROGRAM")
predictions = model.predict(bps1)

for i in range(len(predictions)):
	print(str(bps1[i]) + " => " + str(predictions[i]) + " => " + str(np.round(predictions[i])))
```
